```python
import pandas as pd

with open('sat.trn') as f:
    data = f.readlines()

training_data =pd.DataFrame([list(map(int, line.split())) for line in data])

training_data

# training_data.columns = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y',
#                           'z','aa','ab','ac','ad','ae','af','ag','ah','ai','aj','ak']
```

```python
with open('sat.tst') as f:
    data = f.readlines()

test_data =pd.DataFrame([list(map(int, line.split())) for line in data])

test_data

# test_data.columns = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z',
#                'aa','ab','ac','ad','ae','af','ag','ah','ai','aj','ak']
```

```python
merge_data = pd.concat([training_data, test_data], axis= 0)

merge_data.columns = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z',
               'aa','ab','ac','ad','ae','af','ag','ah','ai','aj','class']

merge_data
```

```python
x = merge_data.iloc[:, :-1]
x
```

```python
y = merge_data.iloc[:,-1]
y
```

## split the dataset in train  &  test set

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 22)
```

# Normalize the data using any normalization technique

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)

x_test = scaler.transform(x_test)
```

# SVC  classification

```python
from sklearn.svm import SVC
```

```python
classification = SVC(kernel= 'linear', random_state= 0)

classification.fit(x_train, y_train)
```

```python
y_pred = classification.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
metrix = confusion_matrix(y_test, y_pred)
metrix
```

```python
score = accuracy_score(y_test, y_pred)
score
```

# Hyper parameter tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = [{'C' : [0.1,0.6,1,10], 'kernel' : ['linear']},
             
             {'C': [1,5,10,12], 'kernel': ['rbf'], 'gamma':[0.01,0.05,0.1,0.2,0.3]}]
```

```python
grid_search = GridSearchCV(estimator= classification, param_grid= parameters, scoring= 'accuracy', cv = 10, n_jobs= -1)

grid_search.fit(x_train, y_train)
```

```python
accuracy = grid_search.best_score_

accuracy
```

```python
grid_search.best_params_
```

### apply the best outcome parameters in SVC

```python
from sklearn.svm import SVC

classification = SVC(C = 5, kernel= 'rbf', gamma= 0.2)
```

```python
classification.fit(x_train, y_train)
```

```python
y_pred2 = classification.predict(x_test)
```

```python
cm = confusion_matrix(y_pred2, y_test)

cm
```

```python
sc = accuracy_score(y_test, y_pred2)                 ## see the accuracy how it has increase

sc
```

# RandomForesst & DecisionTree classification

```python
from sklearn.model_selection import cross_val_score

from sklearn.tree import DecisionTreeClassifier

# Create instances of the other classification models
model1 = DecisionTreeClassifier()

scores_model1 = cross_val_score(model1, x_train, y_train, cv=5, scoring='accuracy')

print("Mean Accuracy (Model 1):", scores_model1.mean())
```

```python
from sklearn.ensemble import RandomForestClassifier

model2 = RandomForestClassifier()

scores_model2 = cross_val_score(model2, x_train, y_train, cv=5, scoring='accuracy')

print("Mean Accuracy (Model 2):", scores_model2.mean())
```
