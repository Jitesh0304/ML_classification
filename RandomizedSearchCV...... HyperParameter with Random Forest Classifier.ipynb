```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Social_Network_Ads.csv')

df.head()
```

```python
gender = pd.get_dummies(df['Gender'], drop_first= True)

gender.head()
```

```python
df1 = pd.concat([df,gender], axis = 1)
df1.head()
```

```python
df1.drop(['User ID','Gender'], axis= 1, inplace= True)
```

```python
df1.head()
```

```python
x = df1[['Age','EstimatedSalary','Male']]
x.shape
```

```python
y = df1['Purchased']
y.shape
```

```python
x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.25, random_state= 0)
```

### Feature scaling.... Standard_scaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)

x_train
```

```python
x_test = scaler.transform(x_test)

x_test
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
r_classifier = RandomForestClassifier(n_estimators= 10, criterion= 'entropy', random_state= 50)
                                   ## n_estimators => it means the number of Decision Tree we required
                                 ## criterion => entropy or gini_impurity
```

```python
r_classifier.fit(x_train, y_train)
```

```python
y_pred = r_classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
score = accuracy_score(y_test, y_pred)

score
```

# RandomizedSearchCV....... HyperParameter tunning

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
from scipy.stats import randint
```

```python
est = RandomForestClassifier(n_jobs= -1)
```

```python
forest_para = {'max_depth': [3,5,10,None], 
               'n_estimators': [10,20,50,100,200,300,500], 
               'max_features' : randint(1,2), 
              'criterion': ['gini', 'entropy'], 
               'bootstrap':[True, False],
               'min_samples_leaf' : randint(1,4)}
```

```python
def hypertuning_rscv(est, para_dict, num_iter, x, y):
    
    random_search = RandomizedSearchCV(est, 
                                       param_distributions= para_dict, 
                                       n_jobs= -1,
                                       n_iter= num_iter, cv =10)
    
     #CV = Cross-Validation ( here using Stratified KFold CV)
    
    random_search.fit(x,y)
    
    hyper_para = random_search.best_params_
    
    hyper_score = random_search.best_score_
    
    return hyper_para, hyper_score
```

```python
rforest_para, rforest_score = hypertuning_rscv(est, forest_para, 40, x, y)
```

```python
rforest_para
```

```python
rforest_score                   ## this is the RandomSearchCV accuracy score
```

### Use the above parameters in the random forest classifier

```python
classifier = RandomForestClassifier(n_estimators= 200,
                                    bootstrap= True,
                                    criterion= 'gini',
                                    max_depth= 3,
                                    max_features= 1,
                                    min_samples_leaf= 2)
```

```python
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
metrix = confusion_matrix(y_test, y_pred)

metrix
```

```python
a_score = accuracy_score(y_test, y_pred)

a_score
```
